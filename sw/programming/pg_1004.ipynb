```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.groupby('sex').mean()
```

```python
tips.groupby('smoker').mean()
```

```python
tips.groupby(['sex', 'size']).mean()
```

```python
tips.groupby(['sex', 'size']).mean().plot()
```

```python
x = tips.groupby(['sex'])[['tip']].mean()
```

```python
x.plot()
```

```python
m  = tips[tips.sex == "Male"]['tip'].values
f = tips[tips.sex == "Female"]['tip'].values

mm = m.mean()
ff = f.mean()
```

```python
plt.bar([0 , 1], [mm, ff])
```

```python
plt.bar([0, 1], [mm, ff], 0.5)
plt.title('타이틀')
plt.xlabel('x라벨')
plt.ylabel('y라벨')
```
