### Closure
- higher order function 
  - 함수를 인자로 받고 함수를 뱉을 수 있는 것

```python
a = 1
def y() :
    a = 2 # 글로벌과 로컬 사이에 있는 변수 : enclosing 변수
    def x():
        a = 3
        return a
    return x()

```

```python
def x(m):
    def y(n):
        return m + n 
    return y

x(3) # x 함수를 사용하면서 y함수가 나온다
x(3)(4) #7 첫번쨰 함수의 인자 값이 함수 내에세 계속 살아있다.

m = x(3) #m은 3을 더해주는 함수가 되는 것
m(4) #7 두가지의 조합으로 유연하게 함수 사용이 가능하다.
```

```python
# 객체 지향에서의 closure 예시 
class A:
    def __init__(self, m) :
        self.m = m
    def __call__(self, n) :
        return self.m + n

model = A(1)
model(4) # 5
```

### Encapsulation (캡술화)



- 파이썬에서 definition을 declaration으로 사용한다. (정의 = 선언) -> 재사용하기 위해서
- 재사용하기 위해서 이름을 이용해 encapsulation을 한다
- 파이썬에서 encapsulation은 2가지 : Class, function
  - 파이썬은 외부에서 내부의 class로 접근할 수 있다. 
  - function은 외부에서 내부로 접근할 수 없다. (Information Hiding), 그러나 접근 통로는 있다(=Interface)
    - Descriptor (meta, decorator)
  - 명명 규칙 : How 보다 What을 중시한다. 

- 재사용하기 위한 함수들을 만들어 사용하는 방식을 함수형 프로그래밍이라 한다.
