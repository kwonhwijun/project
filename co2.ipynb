```python
import numpy as np
import pandas as pd
import patsy
import statsmodels.api as sm
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

x = house_data.average_temp.values
y = house_data.average_power.values

# k-fold 설정
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 후보 노트 위치 설정 (예: 10%에서 90%까지 10% 간격)
min_x, max_x = np.min(x), np.max(x)
candidate_knots = [np.percentile(x, i) for i in range(10, 100, 10) if min_x < np.percentile(x, i) < max_x]

best_mse = float('inf')
best_knot = None

for knot in candidate_knots:
    mse_list = []

    for train_index, test_index in kf.split(x):
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]

        transformed_x_train = patsy.dmatrix("bs(train, knots=[knot], degree=3, include_intercept=False)",
                                             {"train": x_train}, return_type='dataframe')

        model = sm.GLM(y_train, transformed_x_train).fit()
        
        transformed_x_test = patsy.dmatrix("bs(train, knots=[knot], degree=3, include_intercept=False)",
                                            {"train": x_test}, return_type='dataframe')

        predictions = model.predict(transformed_x_test)
        mse = mean_squared_error(y_test, predictions)
        mse_list.append(mse)

    avg_mse = np.mean(mse_list)
    
    if avg_mse < best_mse:
        best_mse = avg_mse
        best_knot = knot

print(f"Best Knot: {best_knot}, MSE: {best_mse}")
```
