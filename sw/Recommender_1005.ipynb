## Data Load

```python
import pandas as pd
```

```python
user = pd.read_csv('ml-1m/users.dat', sep="::", header=None, names=['UserID','Gender','Age','Occupation','Zip-code'])
```

```python
rating = pd.read_csv('ml-1m/ratings.dat', header=None, sep='::')
```

```python
movie = pd.read_csv('ml-1m/movies.dat', sep='::', encoding='latin1', header=None)
```

```python
movie.rename({0:'MovieID', 1:'Title', 2:'Genres'}, axis=1, inplace=True)
```

### Join technique

```python
data = rating.merge(movie).merge(user)
```

```python
data
```

### Reshape - pivot

```python
user_item = data.pivot(index='UserID', columns='MovieID', values='Rating').fillna(0)
# rating.set_index(['UserID', 'MovieID']).unstack() # 동일
```

## Relationship 
- relationship 구하려면 행열 위치 바껴야 됨 

```python
user_item.T
```

```python
temp = user_item.T.corr()
```

```python
temp
```

```python
temp.loc[1005].nlargest(10)
```

```python
temp.loc[1005].sort_values(ascending=False)[1:5] # 내림차순으로 정렬 # 자기 자신은 제외하고 가장 유사도가 높은 아이디 추출 
```

```python
def sim_user(user, n):
    return temp.loc[user].sort_values(ascending=False)[1:n] # 비슷한 영화를 본 경우 추출 # 나와 가까운 몇 명
```

```python
sim_user(1, 5)
```

```python
rating[rating.UserID==1] # 특정 유저가 본 영화만 검색
```

```python
def user_seen(user):
    return rating[rating.UserID==user]['MovieID'].values # 특정한 한 명이 본 영화 
```

```python
user_seen(1)
```

```python
def sim_user_seen(user, n):
    u = temp.loc[user].sort_values(ascending=True)[1:n+1].index
    return rating[rating.UserID.isin(u)]['MovieID'].values
```

```python
sim_user_seen(1,5) # 나와 가까운 5명이 본 영화들 
```

```python
# 가까운 5명이 본 영화중에 내가 본 영화를 제외 # 차집합 개념 이용
set(sim_user_seen(1,5)) - set(user_seen(1))
```

### <span style="background-color:#FFE6E6">위에서 계산한 내용을 함수 하나로 정의</span>

```python
import numpy as np
```

```python
def sim_user_seen_except_me(user, n):
    u = temp.loc[user].sort_values(ascending=True)[1:n+1].index
    um = rating[rating.UserID.isin(u)]['MovieID'].values
    my = rating[rating.UserID==user]['MovieID'].values
    return np.setdiff1d(um, my) # setdiff1d: 차집합 구하기 
```

```python
movie.set_index('MovieID')
```

```python
movie.loc[sim_user_seen_except_me(3, 3)]
```

```python
mi = movie.set_index('MovieID')
```

```python
# 다른 세 명은 봤지만 나는 보지 못한 영화들
mi.loc[sim_user_seen_except_me(3,3)]
```

```python
lake = data.set_index('MovieID')
```

```python
lake.loc[sim_user_seen_except_me(3,3)] # 3번과 가까운 3명이 본 영화들을 추출 
```
